## 01 개념 설명

### 계층적 군집화(Hierarchical Clustering)

1. **비지도 학습 방법**으로, 데이터 포인트들을 계층적 구조로 그룹화하는 방법이다.

2. 이 방법은 데이터를 여러 단계로 (1)나누거나 (2)합쳐가면서 군집을 형성한다.

3. **덴드로그램**으로 데이터의 군집 형성을 시각화하여 이해를 돕는다.

4. 계층적 군집화는 크게 다음과 같은 두 가지로 나뉜다.

    - **병합형 계층적 군집화**(Agglomerative Hierarchical Clustering): 개별 데이터를 시작으로 큰 군집을 형성한다.
    
        - 시작: '각' 데이터 포인트를 개별적인 군집으로 간주한다.

        - 과정: 가장 가까운 두 군집을 합쳐나간다.

        - 종료: 모든 데이터 포인트가 '하나의' 군집이 될 때까지 합친다.
        
        - 특징: 가장 널리 사용되는 방법이다.
    
    - **분할형 계층적 군집화**(Divisive Hierarchical Clustering): 전체 데이터를 큰 군집으로 시작하여 세분화한다.
    
        - 시작: '모든' 데이터 포인트를 하나의 군집으로 간주한다.

        - 과정: 군집을 계속해서 분할한다.

        - 종료: 모든 데이터 포인트가 '개별적인' 군집이 될 때까지 분할한다.

### 덴드로그램(Dendrogram)

1. 계층적 군집화의 결과는 덴드로그램이라는 트리 구조로 시각화할 수 있다.

2. 덴드로그램은 데이터 포인트 간의 군집 형성을 단계별로 보여준다.

### 계층적 군집화의 장단점 #순차적 #계층적

1. 장점
    
    - K-평균 군집화(K-means Clustering)와 달리 군집의 수를 미리 지정할 필요가 없다.

    - 왜? 개체들이 결합되는 순서를 나타내는 트리형태의 구조인 덴드로그램(Dendrogram) 덕분이다.

    - 덴드로그램을 통해 군집 형성 과정을 시각적으로 이해할 수 있다.

2. 단점

    - 계산 비용이 높아 큰 데이터셋에 비효율적일 수 있다.

    - 노이즈나 이상치에 민감할 수 있다.
    
    - 초기 병합 또는 분할의 잘못된 결정이 최종 결과에 큰 영향을 미칠 수 있다.

3. 결론

    - 직관적이나 계산량이 막대하게 된다.

### 어떤 경우에 유용할까?

1. 데이터의 군집 구조를 시각화하고 이해하고자 할 때

2. 데이터 포인트 간의 관계를 탐색할 때
    
3. 소규모 데이터셋에 적용할 때

### 현업에서 어떻게 사용할까? → 비지도 학습과 관련되어 있다.

계층적 군집화는 시각적 이해를 도와주는 덴드로그램을 통해 데이터의 구조를 명확히 파악할 수 있으며, 다양한 산업 분야에서 데이터 분석과 의사결정에 중요한 도구로 사용되고 있다.

1. 마케팅 및 고객 세분화
    - 마케팅 부서에서는 고객을 여러 그룹으로 나누어 세분화할 때 계층적 군집화를 사용한다.
    - 예를 들어, 구매 패턴, 인구 통계학적 데이터 등을 분석하여 비슷한 특성을 가진 고객 그룹을 식별한다.
    - 이를 통해 각 그룹에 맞춤형 프로모션 마케팅 전략을 수립할 수 있다.

2. 생물정보학 및 유전자 분석
    - 계층적 군집화는 생물정보학 분야에서 유전자 발현 데이터를 분석하는 데 사용된다.
    - 비슷한 발현 패턴을 가진 유전자들을 그룹화하여 특정 질병이나 조건과 연관된 유전자를 식별하고, 치료방법을 개발하는 데 유용하다.
    - 덴드로그램을 사용하여 유전자 간의 유사성을 시각화할 수 있다.

3. 문서 분류 및 텍스트 마이닝
    - 문서나 텍스트 데이터를 분류할 때도 계층적 군집화를 사용한다.
    - 예를 들어, 뉴스 기사나 연구 논문을 주제별로 그룹화하여 관련 문서를 쉽게 찾을 수 있다.
    - 이는 문서 간의 유사성을 분석하고 주제별로 분류하는 데 유용하다.

4. 이미지 분류
    - 이미지 데이터를 분석할 때 계층적 군집화를 사용하여 유사한 이미지들을 그룹화할 수 있다.
    - 예를 들어, 대량의 이미지 데이터베이스에서 비슷한 스타일이나 주제를 가진 이미지들을 그룹화하여 검색 효율을 높일 수 있다.

5. 이상치 탐지
    - 계층적 군집화는 데이터셋 내에서 이상치를 탐지하는 데도 사용된다.
    - 비정상적으로 다른 군집에 속하거나 혼자 떨어져 있는 데이터를 식별함으로써 이상치를 찾을 수 있다.
    - 이는 특히 금융, 제조, 사이버 보안 등에서 유용하다.

6. 시계열 데이터 분석
    - 시계열 데이터를 분석할 때 계층적 군집화를 사용하여 비슷한 패턴을 가진 시계열을 그룹화할 수 있다.
    - 예를 들어, 비슷한 계절적 변동을 보이는 제품 판매 데이터를 그룹화하여 재고 관리나 마케팅 전략을 세울 수 있다.

## 02 수식 이해

### 군집 간의 거리 계산 방법

병합형 계층적 군집화에서 군집 간의 거리를 계산하는 방법은 다양하며, 주로 사용되는 방법은 다음과 같다.

1. 최단 연결법 (Single Linkage): 두 군집 사이의 가장 가까운 거리를 기준으로 합친다.

    - 

2. 최장 연결법 (Complete Linkage): 두 군집 사이의 가장 먼 거리를 기준으로 합친다.

    - 

3. 평균 연결법 (Average Linkage): 두 군집 사이의 모든 데이터 포인트 쌍의 평균 거리를 계산한다.

    - 
    
4. 중심 연결법 (Centroid Linkage): 각 군집의 중심 간의 거리를 계산한다.

    -

### K-평균 군집화: 뭐가 문제지? 원핫 인코딩이 문제였다.

```python
X = df.iloc[:,].values
X[:3]
```

```python
# 엘보우 방법을 사용하여 최적의 군집 수를 찾기
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

### 간결하게 만들기

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# KMeans 클러스터링 수행
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# 클러스터 색상 및 레이블 설정
colors = ['red', 'blue', 'green', 'cyan', 'magenta']
labels = ['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4', 'Cluster 5']

# 클러스터 시각화
for i in range(5):
    plt.scatter(X[y_kmeans == i, 0], X[y_kmeans == i, 1], s=100, c=colors[i], label=labels[i])

# 클러스터 중심 시각화
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')

# 그래프 제목 및 축 레이블 설정
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

### 계층적 군집화

```python
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

AgglomerativeClustering 클래스에서 affinity 인자는 metric으로 변경되었다. 따라서 affinity 대신 metric을 사용해야 한다.

```python
from sklearn.cluster import AgglomerativeClustering

# Agglomerative Clustering 수행
hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
