### Naive Bayes Model

Naive Bayes Î∂ÑÎ•òÍ∏∞Î•º ÏÇ¨Ïö©ÌïòÏó¨ Î™®Îç∏ÏùÑ ÌïôÏäµÏãúÌÇ§Í≥†, ÌïôÏäµÏóê ÏÜåÏöîÎêú ÏãúÍ∞ÑÏùÑ Ï¥à Îã®ÏúÑÎ°ú Ï∏°Ï†ïÌïòÏó¨ Ï∂úÎ†•Ìï¥Î≥∏Îã§.

```python
start_time = time.time()
NB = GaussianNB()
NB.fit(x_train,y_train)
end_time = time.time()

print(round(end_time-start_time,2)) # Ïª¥Ìì®ÌÑ∞ÎßàÎã§ Í∞íÏù¥ Îã§Î¶Ñ.
```

1. start_time = time.time(): ÌòÑÏû¨ ÏãúÍ∞ÑÏùÑ Ï¥à Îã®ÏúÑÎ°ú Í∏∞Î°ùÌïúÎã§. Ïù¥Îäî Î™®Îç∏ ÌïôÏäµ ÏãúÏûë ÏãúÍ∞ÑÏùÑ ÏùòÎØ∏ÌïúÎã§.

2. NB = GaussianNB(): Gaussian Naive Bayes Î∂ÑÎ•òÍ∏∞ Í∞ùÏ≤¥Î•º ÏÉùÏÑ±ÌïúÎã§.

    - GaussianNBÎäî Ïó∞ÏÜçÏ†ÅÏù∏ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ Naive Bayes Î∂ÑÎ•òÎ•º ÏàòÌñâÌïòÎäî Îç∞ ÏÇ¨Ïö©ÎêúÎã§.

3. NB.fit(x_train, y_train): x_trainÍ≥º y_train Îç∞Ïù¥ÌÑ∞Î•º ÏÇ¨Ïö©ÌïòÏó¨ Gaussian Naive Bayes Î™®Îç∏ÏùÑ ÌïôÏäµÏãúÌÇ®Îã§.

    - x_trainÏùÄ ÌäπÏÑ±(feature) Îç∞Ïù¥ÌÑ∞Ïù¥Í≥†, y_trainÏùÄ Î†àÏù¥Î∏î(label) Îç∞Ïù¥ÌÑ∞Ïù¥Îã§.

4. end_time = time.time(): Î™®Îç∏ ÌïôÏäµÏù¥ ÏôÑÎ£åÎêú ÌõÑÏùò ÌòÑÏû¨ ÏãúÍ∞ÑÏùÑ Ï¥à Îã®ÏúÑÎ°ú Í∏∞Î°ùÌïúÎã§. Ïù¥Îäî Î™®Îç∏ ÌïôÏäµ Ï¢ÖÎ£å ÏãúÍ∞ÑÏùÑ ÏùòÎØ∏ÌïúÎã§.

5. print(round(end_time-start_time, 2)): Î™®Îç∏ ÌïôÏäµÏóê ÏÜåÏöîÎêú ÏãúÍ∞ÑÏùÑ Í≥ÑÏÇ∞ÌïòÍ≥†, ÏÜåÏàòÏ†ê ÎëòÏß∏ ÏûêÎ¶¨ÍπåÏßÄ Î∞òÏò¨Î¶ºÌïòÏó¨ Ï∂úÎ†•ÌïúÎã§.

```python
NB.score(x_test, y_test)
```
