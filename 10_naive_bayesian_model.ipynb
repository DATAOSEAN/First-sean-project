## 01 개념

### Naive Bayesian Model(단순/순진한 베이즈 모델)

1. 기계 학습과 통계학에서 사용되는 확률 분류기이다.

2. 이 모델은 베이즈 정리를 기반으로 하며, "나이브"라는 이름은 각 특성이 서로 독립적이라고 가정하기 때문에 붙여졌다.

3. 즉, 어떤 특성의 존재가 다른 특성의 존재에 영향을 미치지 않는다고 가정한다.

### 나이브 베이즈 모델의 유래

1. Naive Bayesian Model은 토마스 베이즈(Thomas Bayes)라는 영국 수학자의 이름을 따서 명명된 베이즈 정리(Bayes' Theorem)에서 유래했다.

2. 베이즈 정리는 새로운 증거가 주어졌을 때 어떤 사건이 발생할 확률을 계산하는 방법을 제공한다.

### 수식에 대한 이해

1. P(A∣B)는 사건 B가 주어졌을 때 사건 A가 발생할 확률 (사후 확률) = 알고자 하는 것

2. P(B∣A)는 사건 A가 주어졌을 때 사건 B가 발생할 확률 (우도)

3. P(A)는 사건 A가 발생할 확률 (사전 확률)

4. P(B)는 사건 B가 발생할 확률 (정규화 상수)

Naive Bayesian Model에서는 여러 특성 𝑥1,𝑥2,…,𝑥𝑛이 독립적이라고 가정한다. 따라서, 특정 클래스 𝐶𝑘에 대한 사후 확률은 다음과 같이 계산된다.

하지만, 𝑃(𝑥1,𝑥2,…,𝑥𝑛)는 모든 클래스에 대해 동일하므로, 최대화해야 할 부분만 계산하면 된다.

### 예시를 통해 알아보자:)

예를 들어, 이메일이 스팸인지 아닌지를 분류해보자. 여기에 두 가지 클래스가 있다. 스팸 (S)과 비스팸 (NS).

1. P(S): 스팸일 확률

2. P(NS): 비스팸일 확률

3. P(word∣S): 특정 단어가 스팸 이메일에 등장할 확률

4. P(word∣NS): 특정 단어가 비스팸 이메일에 등장할 확률

단어들이 독립적이라고 가정하고, 이메일이 스팸인지 아닌지를 판단하는 과정은 다음과 같다.

1. 사전 확률 계산: 전체 이메일 중 스팸과 비스팸의 비율을 계산한다. (각각의 확률을 구할 거니까)

2. 우도 계산: 스팸 이메일과 비스팸 이메일에서 특정 단어가 등장할 확률을 계산한다.

3. 사후 확률 계산: 새로운 이메일이 주어졌을 때, 그 이메일이 스팸/비스팸일 확률을 계산한다.

### "cheap"이라는 단어가 주어졌을 때 이메일이 스팸일 확률은?

1. P(A∣B)는 사건 B가 주어졌을 때 사건 A가 발생할 확률 (사후 확률) = cheap이라는 단어가 포함된 새 이메일이 스팸일 확률

2. P(B∣A)는 사건 A가 주어졌을 때 사건 B가 발생할 확률 (우도) = 스팸메일이 주어졌을 때 cheap이 등장할 확률

3. P(A)는 사건 A가 발생할 확률 (사전 확률) = /전체 이메일에서/ 스팸메일이 발생할 확률

4. P(B)는 사건 B가 발생할 확률 (정규화 상수) = cheap이라는 단어가 주어질 확률

1. P(A∣B)는 사건 B가 주어졌을 때 사건 A가 발생할 확률 (사후 확률) = cheap이라는 단어가 포함된 새 이메일이 비스팸일 확률

2. P(B∣A)는 사건 A가 주어졌을 때 사건 B가 발생할 확률 (우도) = 비스팸메일이 주어졌을 때 cheap이 등장할 확률

3. P(A)는 사건 A가 발생할 확률 (사전 확률) = /전체 이메일에서/ 비스팸메일이 발생할 확률

4. P(B)는 사건 B가 발생할 확률 (정규화 상수) = cheap이라는 단어가 주어질 확률

최종적으로 두 확률을 비교하여 더 높은 확률을 가진 클래스를 선택한다.

### 개념 요약

1. Naive Bayesian Model은 특성들이 독립적이라는 가정 하에 베이즈 정리를 사용하는 확률 분류기이다.

2. 베이즈 정리는 사전 확률과 우도를 통해 사후 확률을 계산한다.

3. Naive Bayesian Model은 각 클래스에 대한 사후 확률을 계산하여 가장 높은 확률을 가진 클래스를 선택하는 방식이다.

### 분류될 수 있는 카테고리

1. 지도 학습(Supervised Learning)

    - Naive Bayesian Model은 주어진 레이블이 있는 데이터(훈련 데이터)를 사용하여 학습하는 지도 학습 알고리즘이다.

    - 학습된 모델은 새로운 데이터가 주어졌을 때 그 데이터를 분류하거나 예측하는 데 사용된다. (스팸인지 아닌지)

2. 확률 모델(Probabilistic Models)

    - 이 모델은 확률에 기반하여 작동한다.

    - 데이터가 특정 클래스에 속할 확률을 계산하고, 그 확률을 기반으로 분류를 수행한다.

    - 이는 베이즈 정리를 이용하여 사전 확률, 우도, 사후 확률을 계산하는 방식으로 작동한다.

3. 분류 알고리즘(Classification Algorithms)

    - 대표적인 분류 알고리즘 중 하나이다.

    - 이 모델은 주로 텍스트 분류, 이메일 스팸 필터링, 문서 분류 등 다양한 분류 작업에 널리 사용된다.

4. 통계적 학습(Statistical Learning)

    - 이 모델은 통계적 기법을 이용하여 데이터를 분석하고 예측한다.

    - 데이터의 확률 분포를 학습하고 이를 이용하여 새로운 데이터를 예측하는 방식으로, 통계적 학습의 중요한 예시이다.

5. 특징 독립 가정(Assumption of Feature Independence)

    - 이 모델은 각 특징이 독립적이라는 가정을 바탕으로 작동한다.

    - 즉, 한 특징의 존재가 다른 특징의 존재에 영향을 미치지 않는다고 가정한다.
    
    - 이러한 가정 때문에 "Naive(순진한)"라는 이름이 붙었다.

### 나이브 베이즈 모델을 사용해야 하는 이유

1. 간단하고 빠름

    - Naive Bayesian Model은 간단한 확률 계산을 통해 새로운 데이터를 분류할 수 있다.

    - 계산이 간단하기에 빠르며, 대용량 데이터에서도 효율적으로 동작한다.

    - 구현이 쉬워서 초보자도 이해하고 사용할 수 있다.

2. 효율적인 성능

    - 특히 텍스트 분류(예: 스팸 이메일 분류)에서 매우 좋은 성능을 발휘한다.
    
    - 적은 양의 데이터로도 효과적인 모델을 만들 수 있다.

3. 명확한 확률 해석

    - 모델의 결과를 확률 값으로 제공하기 때문에 해석이 명확하고, 결과의 신뢰성을 평가하기 쉽다.

### 사전 확률, 우도, 사후 확률이란

1. 사후 확률 (Posterior Probability)

    - 주어진 데이터가 있을 때, 그 데이터가 특정 사건에 속할 확률.
    
    - 예를 들어, 특정 단어들이 포함된 이메일이 스팸일 확률.

2. 우도 (Likelihood)

    - 스팸 메일 중에 특정 단어가 있을 확률.

3. 사전 확률 (Prior Probability)

    - 어떤 사건이 일어나기 전에 우리가 그 사건이 발생할 가능성을 알고 있을 때의 확률.

    - 예를 들어, 전체 이메일 중 스팸 이메일의 비율을 알고 있다면, 그것이 스팸 이메일의 사전 확률이다.

### 예시 설명

1. 사후 확률 계산

    - 새로운 이메일에 "할인"이 포함되어 있을 때, 그 이메일이 스팸일 확률을 계산한다.
    
    - 이를 위해 사전 확률과 우도를 곱한 뒤, 모든 가능성을 고려하여 정규화한다.

2. 사전 확률과 우도 곱하기

    - 예를 들어, 전체 이메일 중 30%가 스팸이라면, 스팸의 사전 확률은 0.3이다.

    - 우도는 스팸 이메일 중에서 "할인"이라는 단어가 포함된 이메일의 비율.
    
    - 만약 스팸 이메일의 80%가 "할인"이라는 단어를 포함하고 있다면, 우도는 0.8이다.

### 궁금한 점

1. 각 특성들이 독립적이라는 가정은 무엇을 의미하나?

    - 간단화: 현실에서는 특성들이 상호 의존적일 수 있지만, 독립 가정을 통해 계산을 단순화할 수 있다.
    
    - 계산 용이성: 독립 가정을 통해 각각의 특성을 따로따로 확률을 계산하고 곱할 수 있다. 이를 통해 전체 확률을 쉽게 구할 수 있다.

2. 나이브 베이즈 모델은 각 특성이 독립적이라고 가정하지만, 실제로 특성들이 독립적이지 않은 경우는?

    - 특성들이 독립적이지 않은 상황도 많다.
    
    - 해당 경우에는 특성 선택, 차원 축소, 베이즈 네트워크와 같은 방법을 통해 문제를 보완할 수 있다.

3. 해결 방법 더보기

    - 특성 선택(Feature Selection): 상관된 특성들을 제거하거나 중요한 특성들만 선택하여 모델에 포함시킨다.

    - 차원 축소(Dimensionality Reduction): PCA(주성분 분석)나 LDA(선형 판별 분석)와 같은 기법을 사용하여 상관된 특성들을 결합해 독립적인 주성분으로 변환한다.

    - 베이즈 네트워크(Bayesian Network): 각 특성 간의 종속 관계를 명시적으로 모델링하여 보다 정교한 확률 계산을 수행할 수 있다. 이는 나이브 베이즈 모델보다 복잡하지만, 더 정확한 예측을 가능하게 한다.

    - 다른 모델 사용: 랜덤 포레스트, 서포트 벡터 머신(SVM), 신경망 등 다른 머신러닝 알고리즘을 사용하는 것도 하나의 방법이다. 이러한 모델들은 특성 간의 상관성을 더 잘 처리할 수 있다.

4. 예시 상황 설정

    - 예를 들어, 이메일 스팸 필터링 문제에서 "할인"이라는 단어와 "무료"라는 단어가 자주 함께 등장한다면, 이 두 단어는 독립적이지 않은 관계이다.
    
    - 나이브 베이즈 모델은 이를 독립적이라고 가정하기 때문에 실제로는 "할인"과 "무료"가 함께 등장하는 패턴을 잘 못잡아낼 수 있다.
    
    - 이런 경우 PCA를 통해 특성들을 결합하거나, 베이즈 네트워크를 사용하여 특성 간의 관계를 모델링하는 것이 유리할 수 있다.

### Lemmatizing = 원형 복원

자연어에서는 단어가 다양한 형태로 변형될 수 있다. 이러한 형태는 때때로 유용한 정보를 제공하지만, 스팸 필터링에서는 불필요하기에 처리가 필요하다.

1. 어간 추출기 (Stemmers): 어간 추출기는 규칙 기반의 단순한 도구로, 규칙을 사용해 추가적인 요소를 제거한다. 하지만 자연어는 항상 규칙을 따르지 않기 때문에 어간 추출기는 한계가 있다. 예를 들어, 어간 추출기는 시제를 변경하지 못하지만, 원형 복원기는 'learnt'를 'learn'으로 변환할 수 있다. 어간 추출기는 단순하지만 빠르며, 대부분의 자연어에는 원형 복원기가 없지만 어간 추출기는 존재한다.

2. 원형 복원기 (Lemmatizers): 원형 복원기는 사전을 사용해 추가적인 요소를 제거하고 시제를 변경한다. 이들은 정확하게 동작하지만, 개발하는 데 많은 자원이 필요해서 드물다. 또한 사전을 사용하기 때문에 처리 속도가 느리다.

해당 실습에는 NLTK의 WordNet 원형 복원기를 사용할 것이다. WordNet은 방대한 사전과 같다.

```python
# nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
```

```python
words = ["bats","removed","cheers","good","stopped","went","fired","cleaner","beers"]
for word in words:
    print(lemma.lemmatize(word),end=" ")
```

```python
x_lemmatized = [[lemma.lemmatize(word) for word in text] for text in x_tokenized]
```

```python
print(x_lemmatized[0])
```

### Removing Stopwords = 불용어 제거

자연어에는 'will'과 같은 특별한 의미가 없는 단어들이 있다. 예를 들어, 'will'은 항상 시제를 나타내고, 'and'나 'or' 같은 단어들도 그렇다.

시간을 절약하고 모델의 성능을 향상시키기 위해 이러한 단어들을 제거해야 한다.

불용어를 제거하는 여러 가지 방법이 있지만, 실습에서는 NLTK의 불용어 corpus를 사용할 것이다.

이번에 사용할 corpus에는 11개의 자연어에 대한 불용어 목록이 포함되어 있다.

1. from sklearn.feature_extraction.text import CountVectorizer

    - CountVectorizer는 텍스트 데이터를 벡터 형태로 변환하기 위한 도구이다. 각 단어의 빈도수를 기반으로 벡터를 생성합니다.

2. vectorizer = CountVectorizer(max_features=20000)

    - CountVectorizer 객체를 생성한다.

    - max_features=20000은 가장 많이 등장하는 20,000개의 단어만을 사용하여 벡터를 생성하도록 설정한다.

3. x = vectorizer.fit_transform([" ".join(text) for text in x_prepared]).toarray()

    - .toarray()는 희소 행렬(sparse matrix)을 밀집 행렬(dense matrix)로 변환한다.

### Naive Bayes Model

Naive Bayes 분류기를 사용하여 모델을 학습시키고, 학습에 소요된 시간을 초 단위로 측정하여 출력해본다.

```python
start_time = time.time()
NB = GaussianNB()
NB.fit(x_train,y_train)
end_time = time.time()

print(round(end_time-start_time,2)) # 컴퓨터마다 값이 다름.
```

1. start_time = time.time(): 현재 시간을 초 단위로 기록한다. 이는 모델 학습 시작 시간을 의미한다.

2. NB = GaussianNB(): Gaussian Naive Bayes 분류기 객체를 생성한다.

    - GaussianNB는 연속적인 데이터에 대해 Naive Bayes 분류를 수행하는 데 사용된다.

3. NB.fit(x_train, y_train): x_train과 y_train 데이터를 사용하여 Gaussian Naive Bayes 모델을 학습시킨다.

    - x_train은 특성(feature) 데이터이고, y_train은 레이블(label) 데이터이다.

4. end_time = time.time(): 모델 학습이 완료된 후의 현재 시간을 초 단위로 기록한다. 이는 모델 학습 종료 시간을 의미한다.

5. print(round(end_time-start_time, 2)): 모델 학습에 소요된 시간을 계산하고, 소수점 둘째 자리까지 반올림하여 출력한다.

```python
NB.score(x_test, y_test)
```

Gaussian Naive Bayes 모델 NB가 테스트 데이터 x_test와 레이블 y_test에 대해 정확도를 평가한다.

```python
from sklearn.metrics import confusion_matrix
y_pred = NB.predict(x_test)

conf = confusion_matrix(y_pred=y_pred,y_true=y_test)
import seaborn
seaborn.heatmap(conf,annot=True,fmt=".1f",linewidths=1.5)
import matplotlib.pyplot as plt
plt.show()
```

학습된 Gaussian Naive Bayes 모델을 사용하여 테스트 데이터의 예측 결과를 평가하고, 혼동 행렬(confusion matrix)을 시각화한다.

혼동 행렬은 각 클래스의 예측 정확도를 한눈에 볼 수 있게 해주며, 모델의 성능을 평가하는 데 유용하다.

1. from sklearn.metrics import confusion_matrix: 실제 값과 예측 값 간의 혼동 행렬을 계산하는 데 사용한다.

2. y_pred = NB.predict(x_test): x_test에 대한 예측값을 y_pred에 저장한다.

3. conf = confusion_matrix(y_pred=y_pred, y_true=y_test): 예측 값 y_pred와 실제 값 y_test를 사용하여 혼동 행렬을 계산한다.

4. import seaborn: 데이터 시각화를 위한 라이브러리인 seaborn을 불러온다.

5. seaborn.heatmap(conf, annot=True, fmt=".1f", linewidths=1.5)

    - seaborn.heatmap 함수를 사용하여 혼동 행렬을 히트맵 형태로 시각화한다.

    - conf는 시각화할 혼동 행렬 데이터를 담은 변수다.

    - annot=True는 행렬의 각 셀에 숫자를 표시한다.

    - fmt=".1f"는 셀 안의 숫자를 소수점 첫째 자리까지 표시한다.
    
    - linewidths=1.5는 셀 간의 선 두께를 설정한다.

6. import matplotlib.pyplot as plt: 데이터 시각화를 위해 matplotlib.pyplot을 불러온다.

7. plt.show(): 생성된 히트맵을 화면에 표시한다.

### 혼동행렬 히트맵 시각화 코드를 요약하자면

seaborn과 matplotlib을 함께 사용함으로써, seaborn의 간편한 시각화 기능과 matplotlib의 다양한 제어 기능을 동시에 활용했다.

1. confusion_matrix를 사용하여 혼동 행렬을 계산한다.

2. seaborn의 heatmap 함수를 사용하여 혼동 행렬을 히트맵으로 시각화한다.

3. matplotlib의 plt.show()를 사용하여 생성된 히트맵을 화면에 표시한다.
