### EDA

```python
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns

def plot_categorical_distribution_barchart(df, cat_cols):
    plt.style.use("ggplot")
    num_cols = 3
    num_rows = -(-len(cat_cols) // num_cols)  # à¸§à¸´à¸˜à¸µà¸„à¸³à¸™à¸§à¸“ ceiling division
    plt.figure(figsize=(15, num_rows * 5))
    for i, col in enumerate(cat_cols, 1):
        plt.subplot(num_rows, num_cols, i)
        sns.countplot(data=df, x=col, palette='viridis')
        plt.title(f'Distribution {col}', fontsize=16)
        plt.xlabel(col, fontsize=14)
        plt.ylabel('Count', fontsize=14)
        plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plot_categorical_distribution_barchart(df, cat_cols)
```

```python
def plot_categorical_distribution_piechart(df, cat_cols):
    num_cols = min(len(cat_cols), 3)
    num_rows = (len(cat_cols) + num_cols - 1) // num_cols
    plt.figure(figsize=(15, num_rows * 5))
    #loop
    for i, col in enumerate(cat_cols, 1):
        plt.subplot(num_rows, num_cols, i)
        category_counts = df[col].value_counts()
        plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', 
                colors=sns.color_palette('viridis'),
                startangle=140)
        plt.title(f'Distribution {col} (Pie chart)', fontsize=16)
    plt.tight_layout()
    plt.show()
plot_categorical_distribution_piechart(df, cat_cols)
```

```python
def plot_numerical_distribution_histogram(df, num_cols):
    plt.style.use("ggplot")
    num_cols_to_plot = len(num_cols)
    num_cols_layout = min(num_cols_to_plot, 3)
    num_rows_layout = (num_cols_to_plot + num_cols_layout - 1) // num_cols_layout
    plt.figure(figsize=(15, num_rows_layout * 5))
    for i, col in enumerate(num_cols, 1):
        plt.subplot(num_rows_layout, num_cols_layout, i)
        sns.histplot(df[col].replace([float('inf'), -float('inf')], float('nan')), kde=True)
        plt.xlabel(col)
        plt.ylabel('Freq')
    plt.tight_layout()
    plt.show()
plot_numerical_distribution_histogram(df, num_cols)
```
