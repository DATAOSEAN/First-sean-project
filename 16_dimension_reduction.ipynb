## 01 개념

### 🔎차원 축소(dimension reduction)란?

1. 다차원으로 이루어진 데이터 세트의 차원을 축소해 새로운 차원의 데이터 세트를 생성하는 것이다.

2. 쉽게 말하자면, 데이터의 주요 패턴을 유지하면서 데이터의 차원을 줄이는 기술이다.

3. 일반적으로 피처 선택(feature selection)과 피처 추출(feature extraction)로 나뉜다.

4. 전자는 종속성이 강한 피처를 제거하고 주요 피처만 선택하는 것이다: 보여줄 게 완전히 달라진 나

5. 후자는 기존 피처를 저차원의 (데이터를 잘 설명할 수 있는) 중요 피처로 압축해서 추출하는 것이다.

### 🔎대표적인 차원 축소 기법 알고리즘

1. PCA(Principal Component Analysis): 주성분 분석

    - 데이터의 분산을 최대화하는 방향으로 축을 회전시켜 새로운 축(주성분)을 찾는 방법.
    
    - 주성분은 데이터의 가장 큰 분산을 설명하는 방향으로 정렬된다.

    - 데이터의 분산을 최대한 보존한다는 장점이 있으나, 선형 관계만을 고려한다는 단점이 존재한다.

    - 사용 예시: 이미지 압축, 데이터 시각화 (고차원 데이터를 2D 또는 3D로 시각화)

2. LDA(Linear Discriminant Analysis): 선형 판별 분석법

    - 데이터 분포 간의 차이를 최대화하고 클래스 내 분포를 최소화하는 방향으로 축을 찾는 방법.

    - LDA는 서로 다른 클래스의 데이터를 잘 구분할 수 있는 축을 찾는다.
       
    - 주로 분류 작업에서 사용된다.

    - 장점: 클래스 분리도가 높은 차원을 찾는 데 유용하고, 분류 모델의 성능을 향상시킬 수 있음.

    - 단점: 클래스 간 분산이 적으면 효과가 떨어지고, 데이터가 정규분포를 따른다는 가정이 있음.

    - 사용 예시: 얼굴 인식, 텍스트 분류

3. SVD(Singular Value Denomposition): 특이값 분해

    - 행렬을 세 개의 행렬(왼쪽 특이 벡터, 특이값, 오른쪽 특이 벡터)의 곱으로 분해하는 방법. SVD를 통해 행렬의 차원을 줄일 수 있다.
    
    - 데이터의 구조를 파악하고 차원을 줄이는 데 사용된다.

    - 장점: 데이터의 근사와 압축에 유용하고, 다양한 응용 분야(예: 이미지 압축, 추천 시스템)에 사용.

    - 단점: 계산이 복잡하고 시간이 많이 걸릴 수 있고, 큰 데이터 세트에서는 비효율적일 수 있음.

    - 사용 예시: 추천 시스템 (Netflix 추천 알고리즘), 문서 검색, 이미지 압축

4. NMF(Non-Negative Matrix Factorizition): 비음수 행렬 분해

    - 모든 값이 비음수인 행렬을 두 개의 비음수 행렬의 곱으로 분해하는 방법. NMF는 데이터의 양의 부분만 고려한다.

    - 모든 행렬 원소가 음수가 아닌 행렬로 분해하는 기법.

    - 비선형 구조를 반영할 수 있으며, 주로 텍스트 분석과 이미지 분석 등에서 사용된다.

    - 장점: 데이터의 해석이 직관이고, 텍스트나 이미지 데이터에서 유용.

    - 단점: 최적화 과정이 복잡할 수 있고, 초기화에 민감하여 결과가 달라질 수 있음.

    - 사용 예시: 이미지 분석 (특징 추출), 음성 신호 처리

### 💡차원 축소 알고리즘 요약

각 기법은 데이터의 성질과 분석 목적에 따라 다르게 사용된다.

- PCA: 데이터의 분산을 최대한 보존하는 축을 찾는 기법.

- LDA: 클래스 간 분리도를 최대화하는 축을 찾는 기법.

- SVD: 행렬을 (=특이값을 기준으로) 분해하여 차원을 축소(=정보 압축)하는 기법.

- NMF: 비음수 행렬로 데이터를 분해하여 해석 가능한 구성 성분을 찾는 기법. (=비음수 제약을 통해 직관적인 해석을 제공함.)

### 🔎차원 축소가 필요한 이유

1. 수백 개 이상의 피처로 구성된 데이터 세트의 경우 상대적으로 적은 차원에서 학습된 모델보다 예측 신뢰도가 떨어진다.

2. 차원이 많을 수록 차원 간의 상관관계가 높을 가능성이 커진다. (선형회귀에서 다중 공선성의 문제로 예측 성능 저하)

3. 3차원 이하의 차원 축소를 통해 직관적으로 데이터를 해석할 수 있다.

4. 학습 데이터의 크기도 줄어들어 학습에 필요한 처리 능력을 줄일 수 있다.

### 🔎사용 범위

1. 많은 차원을 가진 이미지나 텍스트에서 차원 축소를 통해 잠재적인 의미를 찾는다.

2. 텍스트 문서의 숨겨진 의미를 추출한다. → PCA, SVD, NMF

3. 문서 내 단어들의 구성에서 숨겨져 있는 시맨틱이나 토픽을 잠재 요소로 간주하기 때문이다. → SVD, NMF

### EDA

```python
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns

def plot_categorical_distribution_barchart(df, cat_cols):
    plt.style.use("ggplot")
    num_cols = 3
    num_rows = -(-len(cat_cols) // num_cols)  # วิธีคำนวณ ceiling division
    plt.figure(figsize=(15, num_rows * 5))
    for i, col in enumerate(cat_cols, 1):
        plt.subplot(num_rows, num_cols, i)
        sns.countplot(data=df, x=col, palette='viridis')
        plt.title(f'Distribution {col}', fontsize=16)
        plt.xlabel(col, fontsize=14)
        plt.ylabel('Count', fontsize=14)
        plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plot_categorical_distribution_barchart(df, cat_cols)
```

```python
def plot_categorical_distribution_piechart(df, cat_cols):
    num_cols = min(len(cat_cols), 3)
    num_rows = (len(cat_cols) + num_cols - 1) // num_cols
    plt.figure(figsize=(15, num_rows * 5))
    #loop
    for i, col in enumerate(cat_cols, 1):
        plt.subplot(num_rows, num_cols, i)
        category_counts = df[col].value_counts()
        plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', 
                colors=sns.color_palette('viridis'),
                startangle=140)
        plt.title(f'Distribution {col} (Pie chart)', fontsize=16)
    plt.tight_layout()
    plt.show()
plot_categorical_distribution_piechart(df, cat_cols)
```

```python
def plot_numerical_distribution_histogram(df, num_cols):
    plt.style.use("ggplot")
    num_cols_to_plot = len(num_cols)
    num_cols_layout = min(num_cols_to_plot, 3)
    num_rows_layout = (num_cols_to_plot + num_cols_layout - 1) // num_cols_layout
    plt.figure(figsize=(15, num_rows_layout * 5))
    for i, col in enumerate(num_cols, 1):
        plt.subplot(num_rows_layout, num_cols_layout, i)
        sns.histplot(df[col].replace([float('inf'), -float('inf')], float('nan')), kde=True)
        plt.xlabel(col)
        plt.ylabel('Freq')
    plt.tight_layout()
    plt.show()
plot_numerical_distribution_histogram(df, num_cols)
```

### 정규화

```python
sub_df = df[['Customer_Age', 'Dependent_count', 'Education_Level', 
            'Marital_Status', 'Income_Category', 'Months_on_book', 
            'Total_Relationship_Count', 'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 
            'Credit_Limit', 'Total_Trans_Amt', 'Total_Trans_Ct']]

sub_df.head()
```

```python
scaler = StandardScaler()
scaler.fit(sub_df)
scaled_sub_df = pd.DataFrame(scaler.transform(sub_df),columns= sub_df.columns )
scaled_sub_df.head()
```

### PCA

1. 주성분 분석(PCA)을 통해 데이터를 3차원으로 축소하고, 그 결과를 3D 산점도로 시각화했다.

2. 각 주성분은 차원을 축소한 새로운 축을 나타내며, 그래프를 통해 데이터의 분포와 구조를 직관적으로 확인할 수 있다.

```python
# 클래스 불러오기: n_components=3은 주성분 3개를 추출하겠다는 의미로, 원본 데이터의 차원을 3차원으로 축소한다.
pca = PCA(n_components=3)
# 데이터 프레임 생성: 각 차원에 "dim1", "dim2", "dim3"이라는 이름을 붙인다.
PCA_df = pd.DataFrame(pca.fit_transform(scaled_sub_df), columns=["dim1", "dim2", "dim3"])
# 그래프 크기 생성
fig = plt.figure(figsize=(10, 8))
# 3D 플롯을 생성하기 위해 add_subplot 메서드를 사용한다. projection='3d'는 3D 그래프를 그리기 위한 설정이다.
ax = fig.add_subplot(111, projection='3d')
# 3차원 산점도를 그린다. PCA_df의 각 주성분("dim1", "dim2", "dim3")을 사용하여 3D 공간에 점을 찍는다.
ax.scatter(PCA_df['dim1'], PCA_df['dim2'], PCA_df['dim3'], c='pink', marker='o')
# 각 축의 레이블을 설정한다.
ax.set_xlabel('Dimension 1')
ax.set_ylabel('Dimension 2')
ax.set_zlabel('Dimension 3')
# 그래프의 제목을 설정한다.
ax.set_title('3D Projection of Data in the Reduced Dimension')
# 그래프를 출력한다.
plt.show()
```

### add_subplot 메서드의 파라미터 알아보기

111은 add_subplot 메서드에서 사용하는 파라미터로, 플롯의 배열과 위치를 지정하는 데 사용된다.

- 1: 전체 플롯 배열에서 행(row)의 수를 나타냅니다.

- 1: 전체 플롯 배열에서 열(column)의 수를 나타냅니다.

- 1: 플롯 배열에서 현재 서브플롯의 인덱스를 나타냅니다.

- 따라서 111은 전체 그래프가 1행 1열로 구성되어 있으며, 그 중 첫 번째(그리고 유일한) 서브플롯을 의미한다.

- 기본적으로 단일 플롯을 생성하는 경우에 자주 사용되는 파라미터 값이다.
